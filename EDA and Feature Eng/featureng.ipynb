# Feature Engineering

```python
import pandas as pd
```

```python
#All the leagues
premier = pd.read_csv("../initial_data/leagues/premier_league_stats.csv")
bundesliga = pd.read_csv("../initial_data/leagues/bundesliga_league_stats.csv")
eredivise = pd.read_csv("../initial_data/leagues/eredivise_league_stats.csv")
ligue1 = pd.read_csv("../initial_data/leagues/ligue1_league_stats.csv")
seriea = pd.read_csv("../initial_data/leagues/seria_league_stats.csv")
laliga = pd.read_csv("../initial_data/leagues/laliga_league_stats.csv")
cl = pd.read_csv("../initial_data/cl/champions_league_stats.csv")
```

Form Score

```python
def form_score(df):
    df["Form Score (Last 5)"] = (df["Won"] * 3 + (df["Played"] - df["Won"] - df["Lost"]) * 1) / df["Played"]
form_score(premier)
form_score(bundesliga)
form_score(eredivise)
form_score(ligue1)
form_score(seriea)
form_score(laliga)
form_score(cl)

premier.head()
```

Home & Away Strength

```python
def strength(df):
    df["Home Strength"] = df["Home Wins"] / df["Home Played"]
    df["Away Strength"] = df["Away Wins"] / df["Away Played"]
strength(premier)
strength(bundesliga)
strength(eredivise)
strength(ligue1)
strength(seriea)
strength(laliga)
strength(cl)

premier.head()
```

Goal Per Point

```python
def goal_per_point(df):
    df["Goal Per Point"] = df["Points"] / df["Goals For"]

goal_per_point(premier)
goal_per_point(bundesliga)
goal_per_point(eredivise)
goal_per_point(ligue1)
goal_per_point(seriea)
goal_per_point(laliga)
goal_per_point(cl)

premier.head()
```

Attacking Strength

```python
def attack(df):
    df["Attacking Strenght"] = df["Goals For"] / df["Played"]

attack(premier)
attack(bundesliga)
attack(eredivise)
attack(ligue1)
attack(seriea)
attack(laliga)
attack(cl)

premier.head()
```

Defense Strength

```python
def defense(df):
    df["Defense Strength"] = df["Goals Against"] / df["Played"]

defense(premier)
defense(bundesliga)
defense(eredivise)
defense(ligue1)
defense(seriea)
defense(laliga)
defense(cl)

premier.head()
```

Normalized Power Score

```python
def normalize_power_score(df):
    df["Normalized Power Score"] = (df["Power Score"] - df["Power Score"].mean()) / df["Power Score"].std()

normalize_power_score(premier)
normalize_power_score(bundesliga)
normalize_power_score(eredivise)
normalize_power_score(ligue1)
normalize_power_score(seriea)
normalize_power_score(laliga)
normalize_power_score(cl)

premier.head()
```
